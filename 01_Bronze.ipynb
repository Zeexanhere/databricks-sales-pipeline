```python
sales_orders_raw = spark.read.format("json").load(
    "/databricks-datasets/retail-org/sales_orders/"
)
print("✅ Data Loaded")
```

```python
%sql
CREATE DATABASE IF NOT EXISTS sales_db
```

```python
(sales_orders_raw
    .write
    .format("delta")
    .mode("overwrite")
    .saveAsTable("sales_db.bronze_sales_orders")
)
print("✅ Bronze table saved!")
```

```python
%sql
SELECT * FROM sales_db.bronze_sales_orders LIMIT 5
```

```python
%sql
DESCRIBE HISTORY sales_db.bronze_sales_orders
```
